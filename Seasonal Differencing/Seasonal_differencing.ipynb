## 1. Understanding the Problem Statement and Dataset
Our data is in a file calld bookings in 'input' directory.

Our results will show differences between the same months of previous years within the series.

We'll take the seasonal difference instead of the first difference because our data is seasonal.

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from math import sqrt
from sklearn.metrics import mean_squared_error

#Importing data
df = pd.read_csv('input/bookings.csv')

#Dropping some columns
df = df.drop(['Year'], axis= 1)
df = df.drop(['Month'], axis= 1)

#Printing head
df.head()
```

## 2. Aggregating dataset at monthly basis



```python
#Aggregating the dataset at monthly level
df.Timestamp = pd.to_datetime(df.Date,format='%m/%Y') 
df.index = df.Timestamp 
df = df.resample('M').mean()
```

Letâ€™s visualize the data to know how it varies over a time period.

This dataset describes the monthly number of bookings over a 10 year period.
The units are 120 observations.

```python
plt.figure(figsize=(12,8))
plt.plot(df.index, df['Bookings'], marker='o')
plt.title("Bookings")
plt.grid('on')
plt.show()
```

## 3. Plotting seasonal differencing
This is the difference between the current and previous seasonal period.

Since our data is monthly, the seasonal difference would be between a given period and 12 periods earlier.

Our results will show differences between the same months of previous years within the series.

```python
import statsmodels.api as sm

series = pd.Series(df['Bookings'])
diff = series.diff(12)
plt.plot(diff)
plt.show()
```
